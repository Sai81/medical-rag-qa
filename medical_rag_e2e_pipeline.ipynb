### Auto merging index

```python
from llama_index.core.node_parser import HierarchicalNodeParser
from llama_index.core.node_parser import get_leaf_nodes
from llama_index.core import StorageContext
from llama_index.core.retrievers import AutoMergingRetriever
from llama_index.core.indices.postprocessor import SentenceTransformerRerank
from llama_index.core.query_engine import RetrieverQueryEngine


def build_automerging_index(
    documents,
    llm,
    embed_model="local:BAAI/bge-small-en-v1.5",
    save_dir="merging_index",
    chunk_sizes=None,
):
    chunk_sizes = chunk_sizes or [2048, 512, 128]
    node_parser = HierarchicalNodeParser.from_defaults(chunk_sizes=chunk_sizes)
    nodes = node_parser.get_nodes_from_documents(documents)
    leaf_nodes = get_leaf_nodes(nodes)
    merging_context = ServiceContext.from_defaults(
        llm=llm,
        embed_model=embed_model,
    )
    storage_context = StorageContext.from_defaults()
    storage_context.docstore.add_documents(nodes)

    if not os.path.exists(save_dir):
        automerging_index = VectorStoreIndex(
            leaf_nodes, storage_context=storage_context, service_context=merging_context
        )
        automerging_index.storage_context.persist(persist_dir=save_dir)
    else:
        automerging_index = load_index_from_storage(
            StorageContext.from_defaults(persist_dir=save_dir),
            service_context=merging_context,
        )
    return automerging_index


def get_automerging_query_engine(
    automerging_index,
    similarity_top_k=12,
    rerank_top_n=6,
):
    base_retriever = automerging_index.as_retriever(similarity_top_k=similarity_top_k)
    retriever = AutoMergingRetriever(
        base_retriever, automerging_index.storage_context, verbose=True
    )
    rerank = SentenceTransformerRerank(
        top_n=rerank_top_n, model="BAAI/bge-reranker-base"
    )
    auto_merging_engine = RetrieverQueryEngine.from_args(
        retriever, node_postprocessors=[rerank]
    )
    return auto_merging_engine
```

```python
from llama_index.core.node_parser import HierarchicalNodeParser
from llama_index.core.node_parser import get_leaf_nodes
from llama_index.core import StorageContext
from llama_index.core.retrievers import AutoMergingRetriever
from llama_index.core.indices.postprocessor import SentenceTransformerRerank
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.storage.docstore import SimpleDocumentStore



def build_automerging_index(
    documents,
    llm,
    embed_model="local:BAAI/bge-small-en-v1.5",
    save_dir="merging_index",
    chunk_sizes=None,
):
    chunk_sizes = chunk_sizes or [432, 144, 48]
    node_parser = HierarchicalNodeParser.from_defaults(chunk_sizes=chunk_sizes)
    nodes = node_parser.get_nodes_from_documents(documents)
    leaf_nodes = get_leaf_nodes(nodes)
    # done till here
    Settings.llm = llm
    Settings.embed_model = embed_model

    docstore = SimpleDocumentStore()
    docstore.add_documents(nodes)
    storage_context = StorageContext.from_defaults(docstore=docstore)

    if not os.path.exists(save_dir):
        automerging_index = VectorStoreIndex(
            leaf_nodes, storage_context=storage_context
        )
        automerging_index.storage_context.persist(persist_dir=save_dir)
    else:
        automerging_index = load_index_from_storage(
            StorageContext.from_defaults(persist_dir=save_dir)
        )
    return automerging_index


def get_automerging_query_engine(
    automerging_index,
    similarity_top_k=12,
    rerank_top_n=6,
):
    base_retriever = automerging_index.as_retriever(similarity_top_k=similarity_top_k)
    retriever = AutoMergingRetriever(
        base_retriever, automerging_index.storage_context, verbose=True
    )
    rerank = SentenceTransformerRerank(
        top_n=rerank_top_n, model="BAAI/bge-reranker-base"
    )
    auto_merging_engine = RetrieverQueryEngine.from_args(
        retriever, node_postprocessors=[rerank]
    )
    return auto_merging_engine
```

```python
from rouge_score import rouge_scorer

# Initialize the ROUGE scorer


reference_texts =[]
generated_texts =[]

def get_and_print_rouge_scores(reference_text_list, generated_text):
  reference_text = " ".join(reference_text_list)
  scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
  reference_texts.append(reference_text)
  generated_texts.append(generated_text)
  scores = scorer.score(reference_text, generated_text)


  rouge1_precision_scores.append(scores['rouge1'][0])
  rouge1_recall_scores.append(scores['rouge1'][1])
  rouge1_f1_scores.append(scores['rouge1'][2])
  rouge2_precision_scores.append(scores['rouge2'][0])
  rouge2_recall_scores.append(scores['rouge2'][1])
  rouge2_f1_scores.append(scores['rouge2'][2])
  rougeL_precision_scores.append(scores['rougeL'][0])
  rougeL_recall_scores.append(scores['rougeL'][1])
  rougeL_f1_scores.append(scores['rougeL'][2])

  return scores['rouge1'][0]

print(reference_texts)
print(generated_texts)
print(rouge1_precision_scores)
print(rouge2_precision_scores)
print(rougeL_precision_scores)
```

### TruLens Metrics

```python
import numpy as np
from trulens.apps.llamaindex import TruLlama
from trulens.core import Feedback
from trulens.providers.openai import OpenAI

def get_prebuilt_trulens_recorder(query_engine, app_id):
  provider = OpenAI()
  # select context to be used in feedback. the location of context is app specific.
  context = TruLlama.select_context(query_engine)

  # Define a groundedness feedback function
  f_groundedness = (
      Feedback(
          provider.groundedness_measure_with_cot_reasons, name="Groundedness"
          )
      .on(context.collect())  # collect context chunks into a list
      .on_output()
      )

  # Question/answer relevance between overall question and answer.
  f_answer_relevance = Feedback(
      provider.relevance_with_cot_reasons, name="Answer Relevance"
      ).on_input_output()

  # Question/statement relevance between question and each context chunk.
  f_context_relevance = (
      Feedback(
          provider.context_relevance_with_cot_reasons, name="Context Relevance"
          )
      .on_input()
      .on(context)
      .aggregate(np.mean)
      )

  standalone = StandAlone()
  f_custom_rouge = (
      Feedback(standalone.rouge_scores, name="ROUGE score")
      .on(reference_text=context.collect())
      .on_output()
      )

  feedbacks = [f_answer_relevance, f_context_relevance, f_groundedness,f_custom_rouge]
  tru_recorder = TruLlama(
      query_engine,
      app_name=app_id,
      app_version = "base",
      feedbacks=feedbacks
  )
  return tru_recorder
```

```python
row = {'current_index':'',
              'rouge1_precison':0,'rouge1_recall':0,'rouge1_fmeasure':0,
              'rouge2_precison':0,'rouge2_recall':0,'rouge2_fmeasure':0,
              'rougeL_precison':0,'rougeL_recall':0,'rougeL_fmeasure':0}

def calculate_average(list_of_strings):
    total_sum = sum(float(item) for item in list_of_strings)

    average = total_sum / len(list_of_strings) if list_of_strings else 0
    return average

#def calculate_average(values):
#  last_10 = [float(value) for value in values[-12:]]
#  if len(last_10) < 12:
#    return None  # Not enough values for the calculation

#  return sum(last_10) / len(last_10)

def aggregate_cumulative_results():
  dummyrow = row.copy()
  dummyrow['current_index'] = current_index_name
  dummyrow['rouge1_precison'] = calculate_average(rouge1_precision_scores)
  dummyrow['rouge1_recall'] = calculate_average(rouge1_recall_scores)
  dummyrow['rouge1_fmeasure'] = calculate_average(rouge1_f1_scores)
  dummyrow['rouge2_precison'] = calculate_average(rouge2_precision_scores)
  dummyrow['rouge2_recall'] = calculate_average(rouge2_recall_scores)
  dummyrow['rouge2_fmeasure'] = calculate_average(rouge2_f1_scores)
  dummyrow['rougeL_precison'] = calculate_average(rougeL_precision_scores)
  dummyrow['rougeL_recall'] = calculate_average(rougeL_recall_scores)
  dummyrow['rougeL_fmeasure'] = calculate_average(rougeL_f1_scores)
  cumulative_results.append(dummyrow)
```
